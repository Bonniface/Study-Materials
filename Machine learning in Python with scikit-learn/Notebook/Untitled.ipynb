```python
amazon_url = "https://consulentidellavoro.mi.it/ordine/gli-iscritti-allordine/"
wanted_list = ['Abagnale Amalia','Numero di iscrizione: 1810','Codice fiscale: BGNMLA60H47A294O',' Attivo',
'Via C. Poma 7Milano - Mi',
'a.abagnale@consulentidellavoropec.it'
]
```

```python
scraper = AutoScraper()
result = scraper.build(amazon_url,wanted_list)
data = scraper.get_result_similar(amazon_url, grouped=True)
print(data)
```

```python
key = list(data.keys())
print(key)
#output
```

```python
scraper.set_rule_aliases({'rule_21e9':'Name', 'rule_sufo':'Numero', 'rule_0fku':'A', 'rule_ikwy':'B', 'rule_9h44':'C', 'rule_msxz':'D', 'rule_mz41':'E', 'rule_occj':'F', 'rule_iba5':'G', 'rule_wyqy':'H', 'rule_vu85':'I', 'rule_v1by':'J', 'rule_8jyy':'K', 'rule_8sts':'L', 'rule_p0uu':'M', 'rule_goj1':'N'})
```

```python
# keeping the same rules for other queries also
scraper.keep_rules(['rule_21e9', 'rule_sufo', 'rule_0fku', 'rule_ikwy', 'rule_9h44', 'rule_msxz', 'rule_mz41', 'rule_occj', 'rule_iba5', 'rule_wyqy', 'rule_vu85', 'rule_v1by', 'rule_8jyy', 'rule_8sts', 'rule_p0uu', 'rule_goj1'])
scraper.save('consulentidellavoro')
```

```python
# specify the path
scraper.load("consulentidellavoro")
result = scraper.get_result_similar("https://consulentidellavoro.mi.it/ordine/gli-iscritti-allordine/", group_by_alias=True)
```

```python
import pandas as pd
df = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in result.items() ]))
# Converting our DataFrame to CSV file
df.to_csv('PlayStation_Data.csv')
```
